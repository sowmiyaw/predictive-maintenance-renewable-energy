```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler



# Remove the timestamp column (not needed for model input)
df = df.drop(columns=['Timestamp'])

# Normalize the features using MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df.drop(columns=['Failure']))  # Ignore the 'Failure' column for scaling

# Re-add the Failure column to the scaled data
df_scaled = np.hstack([df_scaled, df['Failure'].values.reshape(-1, 1)])

# Define the sequence length (e.g., 60 time steps)
sequence_length = 60

# Create sequences and labels
def create_sequences(data, sequence_length):
    sequences = []
    labels = []

    for i in range(len(data) - sequence_length):
        # Get the sequence (X)
        sequence = data[i:i + sequence_length, :-1]  # All features (except 'Failure')

        # Get the label (y) - The 'Failure' at the last time step of the sequence
        label = data[i + sequence_length - 1, -1]  # The last 'Failure' value in the sequence

        sequences.append(sequence)
        labels.append(label)

    return np.array(sequences), np.array(labels)

# Create sequences and labels
X, y = create_sequences(df_scaled, sequence_length)

# Split the data into train, validation, and test sets (70%, 15%, 15%)
train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.15)

X_train, X_val, X_test = X[:train_size], X[train_size:train_size+val_size], X[train_size+val_size:]
y_train, y_val, y_test = y[:train_size], y[train_size:train_size+val_size], y[train_size+val_size:]

# Check the shapes
print("Train set shape:", X_train.shape)
print("Validation set shape:", X_val.shape)
print("Test set shape:", X_test.shape)

# X_train.shape will be: (num_sequences, sequence_length, num_features)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score

# Define the Transformer model
class TransformerModel(nn.Module):
    def __init__(self, input_dim, seq_length, num_heads, num_layers, hidden_dim, output_dim):
        super(TransformerModel, self).__init__()

        self.input_dim = input_dim
        self.seq_length = seq_length
        self.num_heads = num_heads
        self.num_layers = num_layers
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim

        # Input Layer (Linear)
        self.input_layer = nn.Linear(input_dim, hidden_dim)

        # Positional Encoding
        self.positional_encoding = nn.Parameter(torch.zeros(1, seq_length, hidden_dim))

        # Transformer Encoder
        self.encoder_layers = nn.TransformerEncoderLayer(
            d_model=hidden_dim,
            nhead=num_heads,
            dim_feedforward=hidden_dim * 2,
            dropout=0.1
        )
        self.transformer_encoder = nn.TransformerEncoder(
            self.encoder_layers,
            num_layers=num_layers
        )

        # Output Layer
        self.output_layer = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        # Pass through input layer
        x = self.input_layer(x)

        # Add positional encoding
        x += self.positional_encoding

        # Pass through transformer encoder
        x = self.transformer_encoder(x)

        # Take the output of the last time step and pass through output layer
        x = x[:, -1, :]  # Only the last time step's output
        x = self.output_layer(x)

        return x

# Hyperparameters
input_dim = X_train.shape[2]  # Number of features (e.g., 6)
seq_length = X_train.shape[1]  # Sequence length (e.g., 60)
num_heads = 4  # Number of attention heads
num_layers = 2  # Number of transformer layers
hidden_dim = 64  # Hidden dimension size
output_dim = 1  # Output: 1 for binary classification (failure/no failure)

# Instantiate the model
model = TransformerModel(input_dim, seq_length, num_heads, num_layers, hidden_dim, output_dim)

```

```python
# Loss and Optimizer
criterion = nn.BCEWithLogitsLoss()  # Binary Cross-Entropy loss
optimizer = optim.Adam(model.parameters(), lr=1e-3)
```

```python
# Convert data to PyTorch tensors
X_train_tensor = torch.Tensor(X_train)
y_train_tensor = torch.Tensor(y_train).unsqueeze(1)  # Make y a column vector
X_val_tensor = torch.Tensor(X_val)
y_val_tensor = torch.Tensor(y_val).unsqueeze(1)

# Training loop
num_epochs = 20
batch_size = 64

for epoch in range(num_epochs):
    model.train()

    # Training in batches
    for i in range(0, len(X_train_tensor), batch_size):
        X_batch = X_train_tensor[i:i + batch_size]
        y_batch = y_train_tensor[i:i + batch_size]

        # Forward pass
        optimizer.zero_grad()
        outputs = model(X_batch)

        # Calculate loss
        loss = criterion(outputs, y_batch)

        # Backward pass
        loss.backward()
        optimizer.step()

    # Validation
    model.eval()
    with torch.no_grad():
        val_outputs = model(X_val_tensor)
        val_preds = (torch.sigmoid(val_outputs) > 0.5).float()
        val_accuracy = accuracy_score(y_val_tensor, val_preds)

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Val Accuracy: {val_accuracy:.4f}")
```

```python
# Test set
X_test_tensor = torch.Tensor(X_test)
y_test_tensor = torch.Tensor(y_test).unsqueeze(1)

# Evaluate
model.eval()
with torch.no_grad():
    test_outputs = model(X_test_tensor)
    test_preds = (torch.sigmoid(test_outputs) > 0.5).float()
    test_accuracy = accuracy_score(y_test_tensor, test_preds)

print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
# Save the trained model to a file
torch.save(model.state_dict(), 'transformer_predictive_maintenance_model.pt')
print("Model saved as 'transformer_predictive_maintenance_model.pt'")
```

```python
import torch
import torch.nn as nn
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Model class (must match)
class TransformerModel(nn.Module):
    def __init__(self, input_dim, seq_length, num_heads, num_layers, hidden_dim, output_dim):
        super(TransformerModel, self).__init__()
        self.input_layer = nn.Linear(input_dim, hidden_dim)
        self.positional_encoding = nn.Parameter(torch.zeros(1, seq_length, hidden_dim))
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=num_heads, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.output_layer = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = self.input_layer(x)
        x = x + self.positional_encoding
        x = self.transformer_encoder(x)
        x = x[:, -1, :]
        x = self.output_layer(x)
        return x

# Load model
def load_model(model_path):
    input_dim = 5
    seq_length = 60
    model = TransformerModel(input_dim, seq_length, num_heads=4, num_layers=2, hidden_dim=64, output_dim=1)
    model.load_state_dict(torch.load(model_path, map_location=torch.device('cpu')))
    model.eval()
    return model

# Get user input
def get_user_input():
    print("Enter sensor values:")
    temperature = float(input("Temperature (°C): "))
    voltage = float(input("Voltage (V): "))
    current = float(input("Current (A): "))
    vibration = float(input("Vibration (m/s²): "))
    humidity = float(input("Humidity (%): "))
    return np.array([[temperature, voltage, current, vibration, humidity]])

# Main
def main():
    model_path = 'transformer_predictive_maintenance_model.pt'
    model = load_model(model_path)

    input_features = get_user_input()

    scaler = MinMaxScaler()
    scaler.fit(np.array([
        [-50, 0, 0, 0, 0],
        [150, 500, 100, 100, 100]
    ]))
    input_scaled = scaler.transform(input_features)

    sequence = np.repeat(input_scaled, 60, axis=0)
    sequence = sequence.reshape(1, 60, 5)

    X_input = torch.Tensor(sequence)

    with torch.no_grad():
        output = model(X_input)
        prob = torch.sigmoid(output)
        pred = (prob > 0.5).float()

    print("\n--- Prediction Result ---")
    if pred.item() == 1.0:
        print(f"🚨 WARNING: High risk of failure! (Confidence: {prob.item()*100:.2f}%)")
    else:
        print(f"✅ System is healthy! (Confidence: {100 - prob.item()*100:.2f}%)")

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

# 1. Generate dummy dataset
def create_dataset(samples=1000, sequence_length=60, num_features=5):
    X = np.random.rand(samples, sequence_length, num_features)
    y_failure = np.random.randint(0, 2, samples)  # 0 or 1
    y_time_to_failure = np.random.uniform(1, 48, samples)  # 1 to 48 hours
    return X, y_failure, y_time_to_failure

X, y_failure, y_time_to_failure = create_dataset()

# Normalize features
scaler = MinMaxScaler()
X_reshaped = X.reshape(-1, X.shape[-1])
X_scaled = scaler.fit_transform(X_reshaped)
X = X_scaled.reshape(X.shape)

# Train-test split
X_train, X_test, y_failure_train, y_failure_test, y_time_train, y_time_test = train_test_split(
    X, y_failure, y_time_to_failure, test_size=0.2, random_state=42
)

# Convert to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_failure_train = torch.tensor(y_failure_train, dtype=torch.long)
y_failure_test = torch.tensor(y_failure_test, dtype=torch.long)
y_time_train = torch.tensor(y_time_train, dtype=torch.float32)
y_time_test = torch.tensor(y_time_test, dtype=torch.float32)

# 2. Build Transformer model
class TransformerPredictiveMaintenance(nn.Module):
    def __init__(self, feature_size=5, num_heads=2, num_layers=2, hidden_dim=64):
        super(TransformerPredictiveMaintenance, self).__init__()
        self.input_layer = nn.Linear(feature_size, hidden_dim)
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=num_heads, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc_classification = nn.Linear(hidden_dim, 2)  # Binary classification (failure or not)
        self.fc_regression = nn.Linear(hidden_dim, 1)  # Regression (time-to-failure)

    def forward(self, x):
        x = self.input_layer(x)
        x = self.transformer_encoder(x)
        x = torch.mean(x, dim=1)  # Pooling across sequence
        classification_output = self.fc_classification(x)
        regression_output = self.fc_regression(x)
        return classification_output, regression_output.squeeze()

model = TransformerPredictiveMaintenance()

# 3. Set up training
criterion_classification = nn.CrossEntropyLoss()
criterion_regression = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 4. Train the model
num_epochs = 30
batch_size = 32

train_dataset = torch.utils.data.TensorDataset(X_train, y_failure_train, y_time_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for batch in train_loader:
        inputs, labels_failure, labels_time = batch
        optimizer.zero_grad()
        outputs_class, outputs_time = model(inputs)

        loss_classification = criterion_classification(outputs_class, labels_failure)
        loss_regression = criterion_regression(outputs_time, labels_time)

        loss = loss_classification + loss_regression  # Multi-task learning loss

        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}")

# 5. Test the model
model.eval()
with torch.no_grad():
    test_class_output, test_reg_output = model(X_test)
    predicted_failure = torch.argmax(test_class_output, dim=1)

    acc = (predicted_failure == y_failure_test).float().mean()
    mse = nn.MSELoss()(test_reg_output, y_time_test)

print(f"Test Accuracy (Failure Prediction): {acc.item()*100:.2f}%")
print(f"Test MSE (Time to Failure Prediction): {mse.item():.4f}")

# 6. Save the model
torch.save(model.state_dict(), "transformer_predictive_maintenance_model.pt")

print("\n✅ Model trained and saved successfully!")
```
